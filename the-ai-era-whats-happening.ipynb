# The AI landscape

When chatgpt came along a lot of people were mesmerized, got super excited about how well a machine can answer arbitary questions on anything you can think of. However the model was not good at many things, reasoning, math, code etc 
However, eversince the chatgpt was released reseraches and entreprenures are innovating and they found a way to either compensate for model's weaknesses through enabling complementary capablities for the model or via some design patterns. It's amazing how well the ai applications perform now, it was not obvious the ai that just outputs some random text is now able to solve complex, real-world problems tailored to individual needs.
